```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import pandas as pd
```

```python
text = "What fowl grabs the spotlight after the Chinese Year of the Monkey ?"
```

```python
train_df = pd.read_csv('cleaned_trec_train.csv')
test_df = pd.read_csv('cleaned_trec_test.csv')
df = pd.concat([train_df, test_df])
df.head()
```

```python
df.shape
```

# Tokenize

```python
sentences = [word_tokenize(text.lower()) for text in df['text']]
sentences
```

```python
sentences[430]
```

```python
w2v = Word2Vec(sentences, vector_size=300, min_count=1, sg=1, epochs=10)
```

```python
w2v.wv.key_to_index['weakness']
```

```python
vocab_size = len(w2v.wv.index_to_key)
```

```python
w2v.wv.similarity('what', "who")
```

```python
w2v.wv.similarity('thunder', "america")
```

```python
w2v.wv.most_similar(positive=['clouds'])
```

```python
text_encoded = [w2v.wv.key_to_index[word] for word in word_tokenize(text.lower())]
text_encoded
```

```python
text_decoded = [w2v.wv.index_to_key[id] for id in text_encoded]
text_decoded
```

```python
" ".join(text_decoded)
```

```python
text_encoded_embeddings = [w2v.wv.vectors[id] for id in text_encoded]
text_encoded_embeddings
```

```python
len(text_encoded_embeddings)
```

```python
len(text_encoded_embeddings[0])
```
